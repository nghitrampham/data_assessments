## Helper functions 

```python
def evaluation(y_true, y_pred, title = ""):
    """
    This function is used to compute the confusion matrix 
    and other evaluation metrics.
    """
    cm = confusion_matrix(y_true, y_pred)
    print('Confusion Matrix for {}: \n'.format(title), cm)
    total = sum(sum(cm))
    
    print('=============')
    accuracy = (cm[0,0] + cm[1,1]) /total
    print ('Accuracy    : {}'.format(accuracy))
    Specificity = cm[0,0] / (cm[0,0] + cm[0,1])
    print('Specificity :  {}'.format(Specificity))
    sensitivity = cm[1,1] / (cm[1,0] + cm[1,1])
    print('Sensitivity :  {}'.format(sensitivity))
    
    
def plot_swarm(x, y, df):
    sns.set()
    sns.swarmplot(x=x, y = y, data = df)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
def analyte_vs_tl_visualization(df, lot_num = "1"):
    """
    Function used for plotting
    """
    sns.set()
    fig, axs = plt.subplots(ncols=2, figsize=(12, 7))
    sns.swarmplot(x="analyte", y="tl", data = df, ax=axs[0])
    axs[0].set_title("Distribution of test line values for \n \
                    positive and negative samples, lot_num = {}".format(lot_num))

    pos_samples = df[df["analyte"] == "positive"]["tl"]
    neg_samples = df[df["analyte"] == "negative"]["tl"]

    x_ecdf, y_ecdf = ecdf(neg_samples)
    sns.scatterplot(x=x_ecdf, y=y_ecdf, ax=axs[1])
    x_ecdf, y_ecdf = ecdf(pos_samples)
    sns.scatterplot(x=x_ecdf, y=y_ecdf, ax=axs[1])
    axs[1].set_title("ECDF plot of test line values for positive \n \
                    and negative samples, lot_num = {}".format(lot_num))
    axs[1].legend(('positive', 'negative'), loc='lower right')
```

# Model choices:
I think there are several choices to build this model. 

1. The first way is to use logistic regression to fit to the data, then from the trained logistic regression, we can find the threshold value of the explanatory variable (I will discuss this method in detail later) 

2. We can derive the threshold using the ROC curve by picking the optimal value that minimizes the sum of false positive and false negative. 

3. Use a decision tree. A decision tree will minimize the impurity in each leaf node. We can use that threshold that the tree uses to split the node. 

4. Pick threshold based on our plot.

I will proceed with Logistic Regression (and a little bit on Decision Tree) due to time constraints.

# Model Discussion:
1. I think we can use either Decision Tree or Logistic Regression in this case. For logistic regression,
recall that $$\sigma(x) = \frac{1}{1 + e ^-x}$$
The output from logistic regression in our case is defined as : 
  $$\sigma(w^Tx) = \frac{1}{1 + e ^-w^Tx}$$ where x is the explanatory variable (which is test line value), and w is the weight learned from model. 
  Because of the nature of sigmoid function, we have 
  $$
\sigma(w^Tx)=
\begin{cases}
< 0.5 \quad \textit{if} \quad w^Tx < 0\\
> 0.5 \quad \textit{if} \quad w^Tx > 0\\
= 0.5 \quad \textit{if} \quad w^Tx = 0
\end{cases}
$$

Here, 0.5 is also used as a default probability threshold. We can later tune this probability threshold for a better result. For example, suppose we think that a false positive is less costly than a false negative. In that case, we can lower the threshold to 0.3 or so to capture all the positive samples in the trade-off for increasing the false positive rate.   

2. Since we only 1 explanatory variable, we can solve for the equation 
 $$w_{0} + w_{1}x > 0$$ for positive test. This implies 
  $$ \implies x > \frac{-w_{0}}{w_{1}}$$ will result in a positive covid result. Hence, the threshold is $$\frac{-w_{0}}{w_{1}}$$ 
  
 I will go ahead and split train vs. test set. First, we will train to fit the logistic model and decide the threshold, then use the threshold on the test set and evaluate the result. 
 
 Since each lot is independent. I will first train and test on lot 1, then move on to lot 2 and 3

## Lot number 1
- scaling is optional. We split the data to train and test set

```python
### preparing data
X = np.array(lot1_df["tl"])
y = np.array(lot1_df["label"])

### spliting test and train dataset from lot 1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

### fit the model
clf = LogisticRegression(random_state=0).fit(X_train.reshape(-1,1), y_train)
print(clf.coef_, clf.intercept_)
```

```python
threshold = (-clf.intercept_[0]/clf.coef_[0])[0]
threshold ### base threshold without tunning the probability cutoff. 
```

```python
### use the threshold for the test set and evaluate it
y_pred_lot_1 = [1 if i > threshold else 0 for i in X_test]
evaluation(y_test, y_pred_lot_1, "test set, lot 1")
```

```python
### evaluate on the trainning set
evaluation(y_train, [1 if i > threshold else 0 for i in X_train], "training set, lot 1")
```

### Discussion: I think this base model gives a reasonable result. 
We can verify that by looking at the plot of the "tl" value against the label for lot 1. So I re-plot it here for reference. 
We expect that we will get 5 samples of false negative and 3 samples of false positive with that threshold due to the outliner, and some misclassified points near the threshold.
